```python
my_arr = np.random.randint(low=0, high=11, size=(2,2))
print(my_arr)
```

```python
df = pd.DataFrame(data=my_arr, columns=["i","j"])
print(df)
```

```python
df_inv = pd.DataFrame(np.linalg.pinv(df.values), df.columns, df.index)
print(df_inv)
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = [20.00, 15.00]
plt.rcParams["figure.autolayout"] = True

df = pd.read_csv("/Users/pavankumarmvs/Documents/DomainMidtown/IntervalData.csv", header=0)
# display(df.where(df["USAGE_DATE"] >= '04/19/2024').agg({"USAGE_KWH":"sum"}))
ax = df.groupby(df["USAGE_DATE"].map(lambda x: x[:6]+x[-2:])).sum("USAGE_KWH").plot.bar()
ax.bar_label(ax.containers[0])
total = df.where(df["USAGE_DATE"] >= '06/19/2024').agg({"USAGE_KWH":"sum"})
display('Total Usage: KWH',total.values[0],'Avergae: KWH',total.values[0]/df["USAGE_DATE"].nunique(),'Cost: $',total.values[0]*0.117044*1.032)
# display(df.where(df["USAGE_DATE"] >= '06/19/2024').agg({"USAGE_KWH":"mean"}))

df2 = df.groupby([df["USAGE_DATE"], df["USAGE_START_TIME"].map(lambda x: x[:2])]).sum("USAGE_KWH").reset_index()

grouped = df2.groupby(df2["USAGE_DATE"])
finaldf = pd.DataFrame({'USAGE_START_TIME':grouped.get_group(df2["USAGE_DATE"][0])["USAGE_START_TIME"].values})
# display(df["USAGE_DATE"].unique())
for date in df["USAGE_DATE"].unique():
    finaldf[date] = grouped.get_group(date)["USAGE_KWH"].values
# display(finaldf)
finaldf.plot(kind="bar", colormap="viridis")
# plt.savefig("/Users/pavankumarmvs/Documents/DomainMidtown/IntervalData_20240520_20240619.png")
plt.figure()
# finaldf["USAGE_START_TIME"].dtypes
sns.heatmap(finaldf.set_index("USAGE_START_TIME"), annot=True)
plt.show()
```
